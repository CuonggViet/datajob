```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("DataJobs2.csv")
df.head()
```

```python
df.shape
```

```python
null_values_series = ((df.isnull().sum() / df.shape[0]) * 100).round(2) 
null_values_series
```

Let's first see the object data

```python
# Bring only the object columns (strings)
objects = 'object'
object_column_names = df.select_dtypes(include=objects).columns
print(object_column_names, '\n')
```

```python
object_column_names = object_column_names.drop(['Url', 'Job Descriptions'], errors='ignore')
for column in object_column_names:
    print(f'{column}:\n{df[column].value_counts().head()}\n{"-"*50}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in object_column_names:
    plt.figure(figsize=(10,5))
    chart = sns.countplot(data=df, x=column, order = df[column].value_counts().iloc[:10].index)
    plt.title(f'Bar chart for {column}')
    plt.xticks(rotation=45)
    
    # Adding the numbers
    for p in chart.patches:
        chart.annotate(format(p.get_height(), '.0f'), 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', 
                       va = 'center', 
                       xytext = (0, 10), 
                       textcoords = 'offset points')
    plt.show()
```
